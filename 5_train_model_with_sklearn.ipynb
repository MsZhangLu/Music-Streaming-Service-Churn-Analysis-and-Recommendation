### Train model with sklearn

```python
# define function to perform train, test, and get model performance
def train_test_model(clf, X_train, y_train, X_test, y_test):
    # fit model with training data
    clf.fit(X_train, y_train)
    
    # make prediction on training data
    y_train_pred = clf.predict(X_train)
    p_train_pred = clf.predict_proba(X_train)[:,1]
    
    # make prediction on testing data
    y_test_pred = clf.predict(X_test)
    p_test_pred = clf.predict_proba(X_test)[:,1]
    
    # model performance
    get_performance_metrics(y_train, p_train_pred, y_test, p_test_pred)
    plot_roc_curve(y_train, p_train_pred, y_test, p_test_pred)
```

### Calculate matric scores

```python
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score, roc_auc_score
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(y_train, p_train_pred, y_test, p_test_pred):
    roc_auc_train = roc_auc_score(y_train, p_train_pred)
    fpr_train, tpr_train, _ = roc_curve(y_train, p_train_pred)
    
    roc_auc_test = roc_auc_score(y_test, p_test_pred)
    fpr_test, tpr_test, _ = roc_curve(y_test, p_test_pred)
    
    plt.figure()
    lw = 2
    plt.plot(fpr_train, tpr_train, color='green',
            lw=lw, label='ROC Train (AUC = %0.4f)' % roc_auc_train)
    plt.plot(fpr_test, tpr_test, color='darkorange',
            lw=lw, label='ROC Test (AUC = %0.4f)' % roc_auc_test)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()
    
```

```python
import pandas as pd

def get_performance_metrics(y_train, p_train_pred, y_test, p_test_pred, threshold=0.5):
    metric_names = ['AUC', 'Accuracy', 'Precision', 'Recall', 'f1-score']
    metric_values_train = [
        roc_auc_score(y_train, p_train_pred),
        accuracy_score(y_train, p_train_pred>threshold),
        precision_score(y_train, p_train_pred>threshold),
        recall_score(y_train, p_train_pred>threshold),
        f1_score(y_train, p_train_pred>threshold)
    ]
    metric_values_test = [roc_auc_score(y_test, p_test_pred),
                    accuracy_score(y_test, p_test_pred>threshold),
                    precision_score(y_test, p_test_pred>threshold),
                    recall_score(y_test, p_test_pred>threshold),
                    f1_score(y_test, p_test_pred>threshold)
                   ]
    all_metrics = pd.DataFrame({'metrics':metric_names,
                                'train':metric_values_train,
                                'test':metric_values_test},columns=['metrics','train','test']).set_index('metrics')
    print(all_metrics)
```

## Hyperparameter Tunning

### Random Forest
* 3 primary parameters
    * max_features: 
        * max num of features RF is allowed to try in individual tree. 
        * auto: don't put any restrictions (in sklearn auto means sqrt); sqrt: classification
        * increase speed by decreasing max_features
    * n_estimators: 
        * num of trees.
        * higher num of trees give better performance (stronger and more stable), but slower.
    * min_samples_leaf: 
        * a smaller leaf makes the model more prone to capturing noise in train data. 
        * more than 50. 
    
* parameters make the model training easier
    * n_jobs: -1, how many processors is allowed to use
    * random_state': makes a solution easy to replicate. 
    * oob_score: random forest's cross validation

### Gradient Boosting Decision Tree (GBDT)

Boosting algorithms controls both bias and variance, and is considered to be more effective. 

Boosting is a sequential technique which works on the principle of **ensemble**. It combines a set of weak learners, and deliver improved prediction accuracy. At any instant t, the model outcomes are weighted based on the outcomes of hte previous instan t-1. The outcomes predicted correctly are given a lower weight and the ones miss-classified are weighted higher. 

Parameters:
1. Tree-specific parameters: affect each individual tree in the model.
    * min_samples_split: 
        * control overfitting. 
        * higher values prevent a model from learning relations which might be highly specific to the particular sample selected for a tree. 
        * too high values can lead to under-fitting.
    
    * min_samples_leaf: 
        * control overfittng.
        * generally lower values should be chosen for imbalanced class problems.
        * simialr to min_weight_fraction_leaf
    * max_depth:
        * control overfitting.
    * max_leaf_nodes
    * max_features:
        * rule-of-thumb: sqrt(m)
2. Boosting parameters: affect boosting operation in the model.
    * learning_rate:
        * impact of each tree on the final outcome. 
        * lower values are generally preferred.
        * lower values require higher number of trees, and will be computationally expensive
    2. n_estimators:
        * number of sequential trees to be modeled
        * the higher, the more robust the model is. can overfit at a point.
    3. subsample:
        * The fraction of observations to be selected for each tree.
        * Typical values ~0.8 generally work fine but can be fine-tuned further.
3. Miscellaneous parameters: other parameters for overall funcitoning.
    * loss
    2. init
    3. random_state
    4. verbose
    5. warm_start
    6. presort
